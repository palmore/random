```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,101)
# The heat coefficient
hx = x**(-0.1)
# I integrated this by hand to see the following
hx_avg = x**(-0.1)/0.9
```

```python
plt.plot(x,hx,'-k',label=r'$h_x$')
plt.plot(x,hx_avg,'--k',label=r'$\bar h_x$')
plt.title(r'Example 6.1')
plt.xlabel(r'$x$ [m]')
plt.ylabel(r'$h_x, \bar h_x\ \left[\frac{W}{m^2\cdot K}\right]$')
plt.legend()
```
