```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Write the growth factor for several schemes (Exact, First Order Explicit, First Order Implicit, Crank-Nicolson)
def exact(Beta,r):
    return np.exp(-r*Beta**2)

def FOexplicit(Beta,r):
    return (1-2*r)*np.ones_like(Beta)+2*r*np.cos(Beta)

def FOimplicit(Beta,r):
    foo = (1+2*r)*np.ones_like(Beta)-2*r*np.cos(Beta)
    return foo**(-1)

def CN(Beta,r):
    num = np.ones_like(Beta)-r*(np.ones_like(Beta)-np.cos(Beta))
    den = np.ones_like(Beta)+r*(np.ones_like(Beta)-np.cos(Beta))
    return num*den**(-1)
```

```python
beta = np.linspace(-1,1,101)
nplots = 7
r = np.linspace(0,1,nplots)
for k in range(nplots):
    plt.figure(k)
    name = r'$r=$' + str(np.round(r[k],3))
    plt.title(name)
    plt.plot(beta,exact(beta,r[k]),'--k',label='Exact')
    plt.plot(beta,FOexplicit(beta,r[k]),'-',label='1st Order Explicit')
    plt.plot(beta,FOimplicit(beta,r[k]),'-',label='1st Order Implicit')
    plt.plot(beta,CN(beta,r[k]),'-',label='Crank-Nicolson')
    plt.legend()
    plt.xlabel(r'$\beta_m$')
    plt.ylabel(r'$G$')
plt.show()
```
