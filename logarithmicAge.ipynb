```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
```

```python
# Reference values. Feel free to change these.
T = 18 # 18 for adulty-ness or 0.25 for baby-ness
q = 25/18 # 25/18 for adulty-ness or 2 for baby-ness
# Do no change anything below this line.
```

```python
# Life Range represented (in years)
life = np.array([0.25,100])
# Actual Age (in years)
t = np.linspace(life[0],life[1],200)
# Perceived Age
def p(x):
    return np.log(x/T)/np.log(q)
```

```python
# Major ages for babies and adults (in years)
majorAge = np.array([0.25,0.5,1,2,4,5,11,14,18,21,25,35,50,67,75,80])
print('Age : Percieved Age')
for k in majorAge:
    print(k,':',p(k))
```

```python
plt.figure()
# Plot actual age vs percieved age
plt.plot(t,p(t),'-k')
# Terrible Twos
plt.plot([2,2],p(life),'--',label='Age 2: Terrible Two\'s')
# Kindergarten
plt.plot([5,5],p(life),'--',label='Age 5: Kindergarten')
# Middle School
plt.plot([11,11],p(life),'--',label='Age 11: Middle School')
# High School
plt.plot([14,14],p(life),'--',label='Age 14: High School')
# Adulthood
plt.plot([18,18],p(life),'--',label='Age 18: Adulthood')
# Serve in Congress
plt.plot([25,25],p(life),'--',label='Age 25: Serve in Congress')
# Serve as President
plt.plot([35,35],p(life),'--',label='Age 35: Serve as President')
# AARP Membership
plt.plot([50,50],p(life),'--',label='Age 50: Join AARP')
# Retire
plt.plot([67,67],p(life),'--',label='Age 67: Retire')
# Life Expectancy
plt.plot([80,80],p(life),'--',label='Age 80: Life Expectancy (Woman)')
plt.legend()
plt.xlim(life)
plt.xlabel(r'$t\ [Years]$')
plt.ylabel(r'Adulty-ness') # or Baby-ness
```
