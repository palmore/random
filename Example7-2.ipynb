```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Fixed properties for this problem. Don't change this
w = 1 # m
Ts = 230 +273.15 # K
Tinf = 25+273.15 # K
u = 60 # m/s
nu = 26.41*10**(-6) # m2/s
k = 0.0338 # W/(m*K)
Pr = 0.690 
```

```python
# Get the Nusselt at each location
x= np.linspace(0,1,1001)
Re= u*x/nu

# # Nusselt number correlations
Nu_lam = 0.332*(Re**0.5)*Pr**(1/3)
Nu_tur = 0.0296*(Re**0.8)*Pr**(1/3)

# # Heat transfer coefficient
h_lam = k/x*Nu_lam
h_tur = k/x*Nu_tur

# # Critical Re for transition
xc = 500000*nu/u
print(xc)

# # Compute the overall heat transfer coefficient
my_h = (x<xc)*h_lam+(x>=xc)*h_tur

# Plot it
plt.figure()
plt.plot(x,h_lam,'--',label='Laminar')
plt.plot(x,h_tur,'--',label='Turbulent')
plt.plot(x,my_h,'-k',label='Actual h')
plt.xlabel(r'$x\ [m]$')
plt.ylabel(r'$h\ [W/(m^2\cdot K)]$')
plt.title('Exercise 7.2')
plt.legend()
```
