```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Returns amplitude and phase as tuple
def growth(nu,beta):
    G = 1-nu + nu*np.cos(beta) -1j*nu*np.sin(beta)
    amp = np.abs(G)
    phase = np.angle(G)
    return (amp,phase)
    
# def amp(nu,beta):
#     return np.sqrt( (1.0-nu+nu*np.cos(beta))**2+nu**2*np.sin(beta)**2 )
```

```python
M = 100
# After some cancelation, this is beta
beta = 2*np.pi*np.linspace(0,1,M+1)
```

```python
# Let's plot this stuff for various values of nu
NN = 6
nu = (1.0+np.arange(NN))/(NN+1.0)

# Make figure 1 as the amplitude and figure 2 as phase
plt.figure(1)
plt.figure(2)
for k in range(NN):
    name = r'$\nu=$' + str(np.round(nu[k],2))
    (amp,phase) = growth(nu[k],beta)
    plt.figure(1)
    plt.plot(beta,amp,label=name)
    plt.figure(2)
    plt.plot(beta,phase,label=name)
# Show the figures
plt.figure(1)
plt.xlabel(r'$\beta$')
plt.ylabel(r'$|G|$')
ax = plt.gca()
ax.annotate(r'Due to periodicity of cosine, some lines overlap.',xy=(0.25,0.85),xycoords='figure fraction',fontsize=10)
plt.legend()
plt.figure(2)
plt.xlabel(r'$\beta$')
plt.ylabel(r'$\phi$ [Radians]')
plt.legend()
plt.show()
```
